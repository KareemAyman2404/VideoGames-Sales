```python
import pandas as pd #Importing Pandas for Data Cleaning
from tkinter import filedialog #Importing Filedialog to choose the Data Source File
import tkinter as tk #GUI of the open file dialog
```

```python
#File Dialog that allows the user to choose the file
root = tk.Tk()
root.withdraw()
dir = filedialog.askopenfilename()
if dir[-3:].lower() == "csv":
    videogames = pd.read_csv(dir, header=0)
elif dir[-4:].lower() == "xlsx":
    videogames = pd.read_excel(dir, header=0)
```

```python
#The Excel file after converting it to Dataframe, let's have a look on the first 5 values
videogames.head(5)
```

```python
#Checking the datatypes for the Data we got
videogames.dtypes
```

```python
#Checking info about the Data we have
videogames.info()

'''
After Checking the results, it seems that we have null values in several columns
That we need to fix so it won't interrupt our calculations
'''
```

```python
#Checking some Statistical Facts about the data
videogames.describe()
```

```python
# Let's see how can we handle the na Values::

# Step 1: Check how many na values we have
videogames.isnull().sum()

#After Checking we found out that the null values exists mostly in:
'''
critic_score    57338
total_sales     45094
na_sales        51379
jp_sales        57290
pal_sales       51192
other_sales     48888
'''
#This will raise errors in calculations when dealing with such things
```

```python
#Step 2: let's delete the common null values between the columns mentions above since we can' eliminate all the nulls or we will literally lose our data

videogames.dropna(subset=["critic_score", "total_sales", "na_sales", "jp_sales", "pal_sales", "other_sales"], how="all", inplace=True)
```

```python
# Now let's check the data after this simple clean:

videogames.isna().sum()

#we got 4 na values in the developer so we can get rid of that as well

videogames.dropna(subset="developer", inplace= True)
```

```python
#Now let's have another look
videogames.isna().sum()
```

```python
#We will now convert the img column to the image url.

videogames["img"] = "https://www.vgchartz.com" + videogames["img"]

#Now let's have another look
videogames.head(5)
```

```python
#Now let's check the duplicates
print(videogames.duplicated().sum()) #We have no duplicate values as it shows Zero
''' videogames.drop_duplicates(inplace=True) '''  #if we had duplicate values we would use that
```

```python
#Last Update Column isn't necessary for me in this particular dataset and also it includes around 17k na values so it won't be accurate in analysis
videogames.drop("last_update", axis=1, inplace=True)
```

```python
#Let's Check Duplicates once more
videogames.duplicated().sum() #It gives us zero so it's fine and we can now get back to the na values
```

```python
#Now let's see what will we do regarding the Na Values (replace them with mean or 0) so let's have another look at them
videogames.isna().sum()

videogames.describe()

#We could replace the na values with the mean but it won't be good as if a game is 7.2 Score and its sales aren't much, it won't make sense so let's just replace them with zeroes

videogames.fillna(0, inplace=True)
```

```python
videogames.isna().sum() #Now we have no na values
```

```python
#Deleting the total_sales = 0 as if it doesn't have any sales, it won't be that beneficial in the analysis
videogames.drop(videogames[videogames["total_sales"] == 0].index, inplace=True)
```

```python
#Now let's have a look on the data info to see if it's necessary to do anything to it like changing data
videogames.info() #We can change the datatype of the release date from object to date object by doing the follows:

videogames["release_date"] = pd.to_datetime(videogames["release_date"]) #we convert it to datetime first
videogames["release_date"] = videogames["release_date"].dt.date #then we convert it to date
```

```python
#Let's Index the data by the genre to make it more organized 
#You May ask why the genre and not the developer, I will show you why

developer_list = videogames["developer"].unique().tolist()
genre_list = videogames["genre"].unique().tolist()
print(len(developer_list), len(genre_list)) #we can see that we have 3214 developer vs only 20 genre so we can index them by genre better
```

```python
#Indexing
videogames.set_index("genre", inplace=True)
videogames.head(5) #check the Data
```

```python
#Let's now sort by index

videogames.sort_values(by="genre", ascending=True, inplace=True)
videogames.head(10)
```

```python
#Next Thing we wanna do is that we want to export that to MySQL to perform different type of analysis there
#First we have to reset the index (because it causes an error when importing a text index to MySQL Table, I still don't know how to fix it so I found out it's better to reset it :] )
videogames.reset_index(inplace=True)
videogames.head(5)
```

```python
#importing to Database
import mysql
import mysql.connector
from sqlalchemy import create_engine

database_name = input("Please Select the Database Name: \n") #Inserting the desired Database Name

conn = mysql.connector.connect(username="root", password="root", host="localhost", port="3306") #Connecting to MySQL to execute SQL commands

cursor = conn.cursor()


cursor.execute(f"CREATE DATABASE IF NOT EXISTS {database_name};") #Creating Chosen Database if it doesn't exists, if exists, it won't be created

engine = create_engine(f"mysql+pymysql://root:root@localhost:3306/{database_name}?charset=utf8mb4") #Connects pandas to SQL Server

table_name = input("Please Select the Table Name: \n")
if_exists_choice = int(input("Press 1 to append data to the table \nPress 2 to replace data in the table\n"))

try:
    if if_exists_choice ==1:
        if_exists_choice = "append"
    elif if_exists_choice ==2:
        if_exists_choice = "replace"
except ValueError as e:
    print(f"An error occured {e} \nPlease Choose a Valid Number:\n")

videogames.to_sql(name=table_name, con=engine, if_exists=if_exists_choice, index=False) #Transferring Data to the MySQL Table

```
